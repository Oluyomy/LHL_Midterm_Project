#### Quick Clean of the raw data
- Convert '..' to 0 for analysis
- Convert 'number' object into float 
- Convert time in to datetime object

```python
# Replace blank with 0 to convert data float
df_raw = df_raw.replace('..', 0)

# Convert column 5 to 108 from object to float
#df_raw.iloc[:, 5:108]=df_raw.iloc[:, 5:108].astype(float) 
df_raw.iloc[:, 5:123]= df_raw.iloc[:, 5:123].apply(pd.to_numeric, errors='coerce')


# Convert df["Time"] to datetime object
df_raw["Time"] = pd.to_datetime(df_raw["Time"].astype(str), format="%Y", errors='coerce')
df_raw.head()
```

## Split the main dataframe (df) into smaller data frame for analysis
- The data set included world employment info graphic from the top 50 contributor to world economy (https://www.theglobaleconomy.com/rankings/gdp_share/)
- There are 108 columns of data with 850 rows of information generate byt the top 50 country that contribute to world economy.

#### Create a dataframe of general population
- Create df_population by filter and drop columns in df_raw that contain the word 'male'


```python
# Create df_population by dropping all column with 'male'
df_population=df_raw.drop(columns=[column for column in df_raw.columns if "male" in column])

df_population.info()
```

##### Clean the df_population dataframe

```python
# Remove the _female from the column names
df_population.columns = df_population.columns.str.replace("\(% of GDP\)", "%GDP", regex=True)
df_population.columns = df_population.columns.str.replace("\(annual % growth\)", "annual_growth", regex=True)
df_population.columns = df_population.columns.str.replace("\(constant 2005 US$\)", "USD", regex=True)



# Identify delimiter and remove the name
df_population.columns = df_population.columns.str.split("(", n=1).str[0]
df_population.columns = df_population.columns.str.split("[", n=1).str[0]

df_population.head()
```

#### Create gender data
- Create df_gender with data columns that contain the word 'male' --> use for further spliting into male and female data


```python
# Create a dataframe with key identifier for the data
df_gender=df_raw.iloc[:,0:4]

# Add in filter data from the raw data frame that has 'male' in the title
df_gender=pd.concat([df_gender, df_raw.filter(regex="male")], axis=1)

df_gender.head()
```

##### Create dataframe of job information with female population
- Create dataframe with female population data 
    - Use filter to filter out columns name that contain "female" in the title
    - Clean column name by utilizing delimiter

```python
# Create df_female with data from column 0 to 3
df_female = df_gender.iloc[:, 0:4]

# Create a gender column and fill in female
df_female['Gender']= ["female" for i in range(len(df_female))]

# Filter all the column from df3 that has the word "female in column"
df_female=pd.concat([df_female, df_gender.filter(regex='female')], axis=1)
```

```python
df_female.info()
```

```python
# Rename the column  for easy analysis
df_female.columns = df_female.columns.str.split('(', n=1).str[0]

# Remove the female from the column names
df_female.columns = df_female.columns.str.replace("female", "")

# Remove the _female from the column names
df_female.columns = df_female.columns.str.replace(",", "")

df_female.info()
```
