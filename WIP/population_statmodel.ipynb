```python
# histogram visualization of gender (df) columns (wages_and_salaries_pct and employer_pct  to detect the distribution of these variables.)
#This is done to meet models assumption that data must be either symmetrical or mormally distributed.

import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows = 2, 
                         ncols = 2, 
                         figsize=(14, 10)        
                        )

# Plot 0,0
axes[0,0].hist(df['wages_and_salaries_pct'])
axes[0,0].set_title('Histogram of wages_and_salaries_pct')
axes[0,0].set_xlabel('wages_and_salaries_pct')
axes[0,0].set_ylabel('Counts')

# Plot 0,1
axes[0,1].hist(df['wages_and_salaries_pct'], bins=30, color='green')
axes[0,1].set_title('Histogram of wages_and_salaries_pct')
axes[0,1].set_xlabel('wages_and_salaries_pct')
axes[0,1].set_ylabel('Counts')
axes[0,1].set_xlim(left=0)
#axes[0,1].set_xlim(right=0)

# Plot 1,0
axes[1,0].hist(df['employer_pct'])
axes[1,0].set_title('Histogram of employer_pct')
axes[1,0].set_xlabel('employer_pct')
axes[1,0].set_ylabel('Counts')

# Plot 1,1
axes[1,1].hist(df['employer_pct'], bins=30, color='green')
axes[1,1].set_title('Histogram of employer_pct')
axes[1,1].set_xlabel('employer_pct')
axes[1,1].set_ylabel('Counts')
axes[1,1].set_xlim(left=0)

plt.show()
```

```python
# distribution  of wages and salaries percentage showing it has a near normal distribution.
fig, axes = plt.subplots(2, 2, figsize=(19, 10))

#matplotlib
axes[0,0].hist(df['wages_and_salaries_pct'])
axes[0,0].set_title('[Matplotlib] Histogram of wages_and_salaries_pct')

axes[0,1].boxplot(df['wages_and_salaries_pct'])
axes[0,1].set_title('[Matplotlib] Box-Plot of wages_and_salaries_pct')

#seaborn
sns.histplot(ax=axes[1,0], data=df, x="wages_and_salaries_pct")
axes[1,0].set_title('[Seaborn] Histogram of wages_and_salaries_pct')

sns.boxplot(ax=axes[1,1], data=df, y="wages_and_salaries_pct")
axes[1,1].set_title('[Seaborn] Box-Plot of wages_and_salaries_pct')

plt.show()
```

```python
import statsmodels.api as sm
y = df[['wages_and_salaries_pct']]
X = df[['employer_pct',
                'industry_pct',
                'service_pct',	
                'agriculture_pct',
                'laborforce_pct',
                'advance_education_pct',
                'intermediate_education_pct',
                'basic_education_pct',
                'Laborforce_pop_total',
                'Population_total',
                'Population_density']]
X = sm.add_constant(X)
X.head()
```
